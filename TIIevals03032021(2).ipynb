```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame, read_csv
import pandas as pd 

file = r'eva_empty.ods'
df = pd.read_excel(file)

barWidth = 0.1

map="empty"
metrics=['collisions', 'success', 'time', 'path']

for metric in metrics:

  cadrl_subsampling = {}
  esdf = {}
  spatial_horizon = {}
  teb = {}
  mpc = {}

  #number of velocities
  j=3
  i=0
  while (i < j):
    cadrl_subsampling["cadrl_subsampling0{0}".format(i+1)] = [df[metric +'01'][i], df[metric+'02'][i], df[metric+'03'][i]]
    esdf["esdf0{0}".format(i+1)] = [df[metric +'01'][i+j], df[metric+'02'][i+j], df[metric+'03'][i+j]]
    spatial_horizon["spatial_horizon0{0}".format(i+1)] = [df[metric +'01'][i+2*j], df[metric+'02'][i+2*j], df[metric+'03'][i+2*j]]
    teb["teb0{0}".format(i+1)] = [df[metric +'01'][i+3*j], df[metric+'02'][i+3*j], df[metric+'03'][i+3*j]]
    mpc["mpc0{0}".format(i+1)] = [df[metric +'01'][i+4*j], df[metric+'02'][i+4*j], df[metric+'03'][i+4*j]]
    i += 1

  # Set position of bar on X axis
  r1 = np.arange(3)
  r2 = [x + barWidth for x in r1]
  r3 = [x + barWidth for x in r2]
  r4 = [x + barWidth for x in r3]
  r5 = [x + barWidth for x in r4]

  colorCADRLSS = "tab:purple"
  colorESDF = "tab:blue"
  colorSPATIALHORIZON = "tab:red"
  colorTEB = "tab:orange"
  colorMPC = "tab:green"

  if metric == "success":
    fig, ax = plt.subplots()
    rects1 = ax.bar(r1, cadrl_subsampling["cadrl_subsampling01"], width=barWidth, label='SUB-WP',  color=colorCADRLSS, alpha=0.3, ecolor='black',capsize=2)
    rects2 = ax.bar(r2, esdf["esdf01"], width=barWidth, label='LM-WP',  color=colorESDF, alpha=0.3, ecolor='black',capsize=2)
    rects3 = ax.bar(r3, spatial_horizon["spatial_horizon01"], width=barWidth, label='STH-WP',  color=colorSPATIALHORIZON, alpha=0.3, ecolor='black',capsize=2)
    rects4 = ax.bar(r4, teb["teb01"], width=barWidth, label='TEB',  color=colorTEB, alpha=0.3, ecolor='black',capsize=2)
    rects5 = ax.bar(r5, mpc["mpc01"], width=barWidth, label='MPC',  color=colorMPC, alpha=0.3, ecolor='black',capsize=2)


    rects6 = ax.bar(r1, cadrl_subsampling["cadrl_subsampling02"], width=barWidth,   color=colorCADRLSS, alpha=0.6)
    rects7 = ax.bar(r2, esdf["esdf02"], width=barWidth,  color=colorESDF, alpha=0.6)
    rects8 = ax.bar(r3, spatial_horizon["spatial_horizon02"], width=barWidth,   color=colorSPATIALHORIZON, alpha=0.6)
    rects9 = ax.bar(r4, teb["teb02"], width=barWidth,  color=colorTEB, alpha=0.6)
    rects10 = ax.bar(r5, mpc["mpc02"], width=barWidth,   color=colorMPC, alpha=0.6)


    rects11 = ax.bar(r1, cadrl_subsampling["cadrl_subsampling03"], width=barWidth, color=colorCADRLSS, alpha=1)
    rects12 = ax.bar(r2, esdf["esdf03"], width=barWidth,   color=colorESDF, alpha=1)
    rects13 = ax.bar(r3, spatial_horizon["spatial_horizon03"], width=barWidth,   color=colorSPATIALHORIZON, alpha=1)
    rects14 = ax.bar(r4, teb["teb03"], width=barWidth,  color=colorTEB, alpha=1)
    rects15 = ax.bar(r5, mpc["mpc03"], width=barWidth,   color=colorMPC, alpha=1)

  else:
    fig, ax = plt.subplots()
    rects1 = ax.bar(r1, cadrl_subsampling["cadrl_subsampling01"], width=barWidth, label='SUB-WP',  color=colorCADRLSS, alpha=1, ecolor='black',capsize=2)
    rects2 = ax.bar(r2, esdf["esdf01"], width=barWidth, label='LM-WP',  color=colorESDF, alpha=1, ecolor='black',capsize=2)
    rects3 = ax.bar(r3, spatial_horizon["spatial_horizon01"], width=barWidth, label='STH-WP',  color=colorSPATIALHORIZON, alpha=1, ecolor='black',capsize=2)
    rects4 = ax.bar(r4, teb["teb01"], width=barWidth, label='TEB',  color=colorTEB, alpha=1, ecolor='black',capsize=2)
    rects5 = ax.bar(r5, mpc["mpc01"], width=barWidth, label='MPC',  color=colorMPC, alpha=1, ecolor='black',capsize=2)


    rects6 = ax.bar(r1, cadrl_subsampling["cadrl_subsampling02"], width=barWidth,   color=colorCADRLSS, alpha=0.6)
    rects7 = ax.bar(r2, esdf["esdf02"], width=barWidth,  color=colorESDF, alpha=0.6)
    rects8 = ax.bar(r3, spatial_horizon["spatial_horizon02"], width=barWidth,   color=colorSPATIALHORIZON, alpha=0.6)
    rects9 = ax.bar(r4, teb["teb02"], width=barWidth,  color=colorTEB, alpha=0.6)
    rects10 = ax.bar(r5, mpc["mpc02"], width=barWidth,   color=colorMPC, alpha=0.6)


    rects11 = ax.bar(r1, cadrl_subsampling["cadrl_subsampling03"], width=barWidth, color=colorCADRLSS, alpha=0.3)
    rects12 = ax.bar(r2, esdf["esdf03"], width=barWidth,   color=colorESDF, alpha=0.3)
    rects13 = ax.bar(r3, spatial_horizon["spatial_horizon03"], width=barWidth,   color=colorSPATIALHORIZON, alpha=0.3)
    rects14 = ax.bar(r4, teb["teb03"], width=barWidth,  color=colorTEB, alpha=0.3)
    rects15 = ax.bar(r5, mpc["mpc03"], width=barWidth,   color=colorMPC, alpha=0.3)

  #captions
  caption = ""
  if metric == "path":
    caption = "avg. Path Lenght [m]"
    title = "Path Lenght"
  if metric == "time":
    caption = "avg. Time t.g. [s]"
    title = "Time t.g."
  if metric == "success":
    caption = "Success"
    title = caption
  if metric == "collisions":
    caption = "Collisions"
    title = caption

  # Add xticks on the middle of the group bars
  plt.xlabel('No. Obstacles', fontsize=15)
  plt.xticks([r + 2*barWidth for r in range(3)], ['5', '10', '20'])
  plt.ylabel('{}'.format(caption), fontsize=15)
  plt.title("{} over No. Obstacles".format(title), fontweight='bold', fontsize=16)

  ax.grid('on')
  # Create legend & Showt graphic
  plt.legend(loc='upper left')
  plt.savefig('{0}obs_{1}.png'.format(metric,map))
  plt.show()
```

```
from tabulate import tabulate
from texttable import Texttable

import latextable

rows = [['Rocket', 'Organisation', 'LEO Payload (Tonnes)', 'Maiden Flight'],
        ['Saturn V', 'NASA', '140', '1967'],
        ['Space Shuttle', 'NASA', '24.4', '1981'],
        ['Falcon 9 FT-Expended', 'SpaceX', '22.8', '2017'],
        ['Ariane 5 ECA', 'ESA', '21', '2002']]

table = Texttable()
table.set_cols_align(["c"] * 4)
table.set_deco(Texttable.HEADER | Texttable.VLINES)
table.add_rows(rows)

print('Tabulate Table:')
print(tabulate(rows, headers='firstrow'))

print('\nTexttable Table:')
print(table.draw())

print('\nTabulate Latex:')
print(tabulate(rows, headers='firstrow', tablefmt='latex'))

print('\nTexttable Latex:')
print(latextable.draw_latex(table, caption="A comparison of rocket features."))
```
